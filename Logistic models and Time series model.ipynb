```python
# drops rows having NA values in outcome column
df=df.dropna(subset=['outcome'])
```

```python
# normalise the data from 5th column until end
df.iloc[:,4:]=df.iloc[:,4:].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
# Drop rows where the outcome is 0.135204
df = df[df['outcome'] != 0.135204]
```

```python
df = df[df['outcome'].isin([0, 1])]
```

```python
df['outcome'].value_counts()
```

```python
#Train test split
X = df.drop(['outcome','group','ID'],axis=1)
y = df['outcome']
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train,X_test,y_train,y_test.shape
```

```python
# logistic regression model to predict the outcome
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
# evaluate the model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 score:", f1_score(y_test, y_pred))
```

```python
from prophet import Prophet
import pandas as pd

# Make sure 'Date' is datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# List of drug columns to forecast
drug_columns = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03', 'R06']

# Loop through each drug column
for drug in drug_columns:
    print(f"Processing data for {drug}...")
    # Prepare data for Prophet: select Date and current drug column, rename to 'ds' and 'y'
    ts_data = df[['Date', drug]].rename(columns={'Date': 'ds', drug: 'y'})

    # Create and fit the Prophet model
    model = Prophet()
    model.fit(ts_data)

    # Make future dataframe (e.g., for the next 30 days)
    future = model.make_future_dataframe(periods=30)
    forecast = model.predict(future)

    # Display the last few rows of the forecast for this drug
    print(f"Forecast for {drug}:")
    display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
    print("-" * 30) # Separator for clarity
```
