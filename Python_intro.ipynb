```python
def fn_bmicalculation(weight,height):
  bmi = weight/height**2
  return bmi

def classify_bmi_group (bmi):
 if bmi > 25:
  return "overweight"
 elif bmi < 18:
  return "underweight"
 else:
  return "normal"

bmi1 = fn_bmicalculation(50,1.52)
category1 = classify_bmi_group(bmi1)
print(f"BMI: {bmi1:.2f}, Category: {category1}")

bmi2 = fn_bmicalculation(60,1.65)
category2 = classify_bmi_group(bmi2)
print(f"BMI: {bmi2:.2f}, Category: {category2}")

bmi3 = fn_bmicalculation(70,1.75)
category3 = classify_bmi_group(bmi3)
print(f"BMI: {bmi3:.2f}, Category: {category3}")
```

```python
# Round Brackets () → used in functions and math operations
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)  # Math expression uses ()
    return bmi

# Square Brackets [] → used to store and access list items
weights = [50, 60, 70,90,50,45,55,65,52]  # list of weights
heights = [1.52, 1.65, 1.75,1.65,1.75,1.45,1.25,1.80,1.50]  # list of heights

# Loop to go through each patient and calculate BMI
for i in range(len(weights)):
    bmi = calculate_bmi(weights[i], heights[i])  # Accessing items with []

    # Angle Brackets < > → used in if conditions
    if bmi < 18:
        category = "Underweight"
    elif bmi > 25:
        category = "Overweight"
    else:
        category = "Normal"

    # Output
    print(f"Patient {i+1}: BMI = {bmi:.2f} → {category}")
```
